```python
# q1 (A)
def reverse_str(s):
  return s[::-1]
```

```python
# q2 (B)
def reverse_sen(s):
  newSen = "" 
  splited = s.split(" ")
  for word in splited:
    newSen += reverse_str(word) 
    newSen += " " 
  return newSen
```

```python
# test
print(reverse_str("happy"))
print(reverse_sen("happy little dogs and cat"))
```

```python
# q2
def count(num):
  # print(num / 3, num / 5, 2 * (num / 15))
  num = num - (int(num / 3) + int(num / 5)) + (2 * int(num / 15))
  return num
```

```python
# test 
print(count(15))
```
